```python
import pandas as pd

df = pd.read_csv("weather_data.csv")
df
```

```python
# Using Python Dictionary
weather_data = {
    'day': ['1/1/2023','1/2/2023','1/3/2023'],
    'temperature': [32,35,28],
    'windspeed': [6,7,2],
    'event': ['Rain','Sunny','Snow']
}

df = pd.DataFrame(weather_data)
df
```

```python
# Using Tuples List
weather_data = {
    ('1/1/2023', 32, 6, 'Rain'),
    ('1/2/2023', 35, 7, 'Sunny'),
    ('1/3/2023', 28, 1, 'Snow'),
}

df = pd.DataFrame(weather_data, columns=["day", "temperature", "windspeed", "event"])
df
```

```python
# Using Dictionary List
weather_data = [
    {'day': '1/1/2023', 'temperature': 32, 'windspeed': 6, 'event': 'Rain'},
    {'day': '1/2/2023', 'temperature': 35, 'windspeed': 7, 'event': 'Sunny'},
    {'day': '1/3/2023', 'temperature': 21, 'windspeed': 2, 'event': 'Snow'}
]

df = pd.DataFrame(weather_data)
df
```
