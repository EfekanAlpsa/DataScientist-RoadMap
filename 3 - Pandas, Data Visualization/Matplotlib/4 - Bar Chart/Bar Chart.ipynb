# Matplotlib Tutorial: Bar Chart

```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
company=['GOOGL','AMZN','MSFT','FB']
revenue=[90,136,89,27]
profit=[40,2,34,12]
```

```python
xpos= np.arange(len(company))#It won't honor string values here thats why we did that
xpos
```

```python
plt.bar(xpos,revenue)
```

```python
plt.xticks(xpos,company)  # arange[(0,1,2,3)] = company=['GOOGL','AMZN','MSFT','FB']
plt.ylabel('revenue(bln)')
plt.title('US Tech Stocks')
plt.bar(xpos,revenue,label='Revenue')
plt.bar(xpos,profit,label='Profit')
plt.legend()
```

```python
plt.xticks(xpos,company)  # arange[(0,1,2,3)] = company=['GOOGL','AMZN','MSFT','FB']
plt.ylabel('revenue(bln)')
plt.title('US Tech Stocks')
plt.bar(xpos-0.2,revenue,width=0.4,label='Revenue')
plt.bar(xpos+0.2,profit,width=0.4,label='Profit')
plt.legend()
```

```python
plt.yticks(xpos,company)  # arange[(0,1,2,3)] = company=['GOOGL','AMZN','MSFT','FB']
plt.title('US Tech Stocks')

plt.barh(xpos-0.2,revenue,label='Revenue')
plt.barh(xpos+0.2,profit,label='Profit')
plt.legend()
```
