# Matplotlib Tutorial: Pie Charts

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

## Draw pie chart to track down home expenses

```python
exp_vals = [1400,600,300,410,250]
exp_labels = ["Home Rent","Food","Phone/Internet Bill","Car ","Other Utilities"]
plt.axis("equal")
plt.pie(exp_vals,labels=exp_labels,radius=1.5) # We can increase radius of chart with radius=
plt.show()
```

### Show percentages for every pie. Specify radius to increase chart size

```python
plt.axis("equal")
plt.pie(exp_vals,labels=exp_labels,radius=1.5, autopct='%0.1f%%',shadow=True)
plt.show()
```

### Explode

```python
plt.axis("equal")
plt.pie(exp_vals,labels=exp_labels,radius=1.5, autopct='%0.1f%%',shadow=True,explode=[0,0.5,0,0.1,0.2])
plt.show()
```

### counterclock and angle properties

```python
plt.axis("equal")
plt.pie(exp_vals,labels=exp_labels, shadow=True, autopct='%1.1f%%',radius=1.5,explode=[0,0,0,0.1,0.2],counterclock=True, startangle=45)
plt.show()

# You can rotate your pie chart with startangle=anglevalue
```
