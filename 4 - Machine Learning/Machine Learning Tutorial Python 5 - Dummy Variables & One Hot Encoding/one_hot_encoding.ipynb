```python
import pandas as pd 
```

```python
df = pd.read_csv("homeprices.csv")
df
# There are dummy variables on town column. 
#All values are string and we need to assign them with numbers
```

```python
dummies = pd.get_dummies(df.town)
dummies
```

**concat** method which you can call to join two data frames

It takes an array as an input where you will supply two data frames that you want to join 

You also need to tell how do you want to join. So the axis you want to use is of course columns!

```python
merged = pd.concat([df, dummies], axis='columns')
merged
```

```python
final = merged.drop(['town'], axis='columns')
final
```

## Dummy Variable Trap

When you can derive one variable from other variables, they are known to be multi-colinear. Here if you know values of california and georgia then you can easily infer value of new jersey state, i.e. california=0 and georgia=0. There for these state variables are called to be multi-colinear. In this situation linear regression won't work as expected. Hence you need to drop one column.

**NOTE: sklearn library takes care of dummy variable trap hence even if you don't drop one of the state columns it is going to work, however we should make a habit of taking care of dummy variable trap ourselves just in case library that you are using is not handling this for you**

```python
final = final.drop(['west windsor'], axis='columns')
final
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
X = final.drop('price',axis='columns')
X
```

```python
y = final.price
y
```

```python
model.fit(X, y)
```

```python
model.predict([[2800,0,1]]) # 2800 sqr ft home in robbinsville
```

```python
model.predict([[3400,0,0]]) # 3400 sqr ft home in west windsor
```

It is telling me a 681241 dollars.
3300 square foot home in west windsor was 650000. 

**This result little bit bigger.**

If you want to know how accurate your model is you can use **model.score()** method supply your x and y .

What it will do is will calculate the predicted values for all of the rows in x and then it will compare predictor value with the actual values which is this y. After that it will use some formula to calculate the score .

Here the score is 0.95 which means my model is 95 accurate if the score is 1 it means it is like perfect

```python
model.score(X, y)
```

## Using sklearn OneHotEncoder

First step is to use label encoder to convert town names into numbers

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
dfle = df
dfle.town = le.fit_transform(df.town) # Use label encoder to fit and transform.
dfle # fit & transform means it takes the label column as an input so my label column is town
```

```python
# X is your training data set and y is your dependent variable
X = dfle[['town','area']].values
X
```

```python
y = dfle.price
y
```

What will happen it will assume all my x columns to be categorical variables hence. What you need to do when you create onehotencoder is you need to specify categorical features. 

Later on whatever x I'm supplying the zeroth column in that x is my categorical feature

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
ohe = ColumnTransformer([('town', OneHotEncoder(), [0])], remainder = 'passthrough')
```

```python
X = ohe.fit_transform(X)
X
```

```python
X = X[:,1:]
# Take all the rows and for the columns takes all the columns from index 1 onwards
X
```
